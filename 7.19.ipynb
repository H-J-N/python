## while 循环
- 当一个条件保持真的时候while循环重复执行语句
- while 循环一定要有结束条件，否则很容易进入死循环
- while 循环的语法是：

    while loop-contunuation-conndition:

        Statement

## 示例：
sum = 0

i = 1

while i <10:

    sum = sum + i
    i = i + 1

## 错误示例：
sum = 0

i = 1

while i <10:

    sum = sum + i

i = i + 1
- 一旦进入死循环可按 Ctrl + c 停止

# 验证码
- 随机产生四个字母的验证码，如果正确，输出验证码正确。如果错误，产生新的验证码，用户重新输入。
- 验证码只能输入三次，如果三次都错，返回“别爬了，我们小网站没什么好爬的”
- 密码登录，如果三次错误，账号被锁定


```python
import random
```

```python
a=random.randint(0,9)
b=random.randint(0,9)
c=random.randint(0,9)
d=random.randint(0,9)
print(a,b,c,d)
(a1,b1,c1,d1)=eval(input('a,b,c,d'))
while 1:
    if (a1,b1,c1,d1)==(a,b,c,d):
        print('对了')
        break
    else:
        a=random.randint(0,9)
        b=random.randint(0,9)
        c=random.randint(0,9)
        d=random.randint(0,9)
        print(a,b,c,d)
        (a1,b1,c1,d1)=eval(input('a,b,c,d'))
```

## for 循环
- Python的for 循环通过一个序列中的每个值来进行迭代
- range(a,b,k)， a，b，k 必须为整数
- a: start
- b: end
- k: step
- 注意for 是循环一切可迭代对象，而不是只能使用range

```python
for i in range(10):
    print(i)
```

```python
c=[1,2,3]
c.__iter__
```

```python
for i in c:
    print(i)
```

## 嵌套循环
- 一个循环可以嵌套另一个循环
- 每次循环外层时，内层循环都会被刷新重新完成循环
- 也就是说，大循环执行一次，小循环会全部执行一次
- 注意：
> - 多层循环非常耗时
  - 最多使用3层循环

## 关键字 break 和 continue
- break 跳出循环，终止循环
- continue 跳出此次循环，继续执行

```python
i = 1
while i<=10:
    print('hahaha')
```

```python
for i in range(10):
    if i==5:
        pass
    else:
        print(i)
```

```python
for i in range(10):
    for j in range(10):
        for k in range(10):
            print(i,j,k)
```

```python
for i in range(10):
    for j in range(10):
        for k in range(10):
            if k==5:
                break
            print(i,j,k)
```

```python
for i in range(1, 10):
    for j in range(1, i+1):
        print('{}x{}={}\t'.format(j, i, i*j), end='')
    print()
```

```python
for i in range (1,10):
    for j in range(1,10):
        print(j,"x",i,"=",i*j,"\t",end="")
        if i==j:
            print("")
            break
```
